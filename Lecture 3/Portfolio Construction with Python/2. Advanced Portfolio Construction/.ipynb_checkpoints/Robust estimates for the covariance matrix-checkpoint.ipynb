# Module 2: Robust estimates for the covariance matrix
---

Modern portfolio theory was born with the efficient frontier analysis of Markowitz (1952). Unfortunately, early applications of the technique, based on naïve estimates of the input parameters, have been found to be of little use because they often lead to non-sensible portfolio allocations. 
The focus now is on bridging the gap between portfolio theory and portfolio construction by showing how to generate enhanced parameter estimates so as to improve the quality of the portfolio optimization outputs (optimal portfolio weights), with a focus on risk parameter estimates. We first address sample risk and explain how to improve covariance matrix estimation via the use of factor and/or Bayesian techniques and statistical shrinkage estimators.
We then move on to addressing stationarity risk, namely the fact that risk parameters are not constant but move over time. We start with basic rolling-window and exponentially-weighted moving average analysis and then move on to the estimation of covariance parameters with autoregressive conditional heteroskedasticity and state-dependent models.

# 1. The curse of dimensionality

The so-called curse of dimensionality refers to the fact that the number of parameters to estimate dramatically increases with the number of variables. This is especially true for covariance matrix, in which we need to estimate $\frac{n(n-1)}{2}$ correlations.

How can we adress this problem?

The natural answer is to increase the sample size, either increasing the granularity of the data or its length.
What if the sample size cannot be increased? We can try to reduce the number of parameters to estimate.

## Risk models

## 1. Constant correlation model
One way to do this is to assume a constant correlation parameter for all assets. 
We will then have that the estimated covariance between two assets will be:

$$\hat{\sigma}_{i,j} = \hat{\sigma}_i\hat{\sigma}_j\hat{\rho}$$

The optimal estimator $\hat{\rho}$ is to take the average of correlations of all assets:

$$\hat{\rho} = \frac{1}{N(N-1)}\sum_{i,j = 1}^N \rho_{i,j}$$ for ${i \neq j}$

Why doing this? Given the high noise to signal ratio in a sample, there can be really extreme estimators for some correlations (either very high or too low). The idea is therefore to shrink these estimators to get rid of noise.

Even though this method may look too naive, Elton and Gruber (1973) show that the porfolios found using the constant correlation estimator outperform out-of-sample those portfolios created with the sample-based estimates of the correlations.

## 2. Factor model for the covariance matrix
Using a factor model is a reasonable way of reducing the amount of parameters to estimate. Assume that we have N assets and a smaller number of factors K that explains the variabiliry of returns of those N assets. We can use the factor model to get expressions for the variance and covariance of the stock returns.

Using some mathematical development, we can find that the variance of each stock will be a function of the factors and its corresponding betas. The covariance between two stock returns, on the other hand, will contain a term which is the covariance between the specific risk of each stock (call it $cov(\epsilon_i\epsilon_j$). But trying to estimate that term would make as return to the curse of dimesionality. We are therefore going to impose the assumption that these terms are equal to zero. That is, specific returns are uncorrelated. If the factor model is well specified, this assumption is not so restrictive.

If we manage to find uncorrelated factors, we can therefore significantly reduce the number of parameters necessary to estimate the covariance matrix. Think of a 100 stock universe, and a 5 factor model. We need to find the betas of each stock with respect to each factor, which means we have 500 parameters to estimate (in contrast to the 4,950 correlations).


#### Choice of the factor model

Types of multifactor models:

- Explicit factor models:

    - Macro-factors (inflation, growth, etc)
    - Micro-factors (country, size, P/E, etc)
    
- Implicit factor model 

    - Statistical factors(data driven, e.g. PCA)

An implicit factor model is often preferred since it lets the data tell us what the underlying factors are, un thus alleviating model risk.


## 3. Honey I shrunk the covariance matrix!

Ledoit and Wolf (2003, 2004) propose a shrinkage approach for reducing the estimation error in the covariance matrix by using a weighted average of the sample covariance matrix and another structured estimator. 

Remember that the problem is a trade-off between sample risk and model risk. The sample-based estimates for covariance parameters have a lot of sample risk, because of the high noise to signal ratio. On the other hand, when imposing a methodology such as the constant correlation or factor models, an ad-hoc structure is introduced, and thus there is a fair amount of model risk.

The idea behind the shrinkage method is that these two methodologies can be mixed. They call it the **shrinkage** of the covariance matrix. If properly implemented, this shrinkage could fix the problem of the sample covariance matrix.

The essential idea is that the unbiased but poorly estimated sample covariance can be combined with a structured estimator $F$ (it can be the factor model covariance), using the below formula (where $\delta$ is the shrinkage constant and $S$ is the sample covariance matrix):

$$ \hat{\Sigma} = \delta F +(1-\delta)S$$

It is called shrinkage because it can be thought of as “shrinking” the sample covariance matrix towards the other estimator, which is accordingly called the shrinkage target. The shrinkage target may be significantly biased but has little estimation error. There are many possible options for the target, and each one will result in a different optimal shrinkage constant $\delta$. 

Academic research has shown that out-of-sample performance of the optimized portfolios based on this shrinkage estimate for the covariance matrix, is actually better than the out-of-sample performance of the optimized portfolio based on either one of the two ingredients, the $F$ covariance matrix or the $S$ covariance matrix. 


Jagannathan and Ma (2003) show that imposing constraint on weights is equivalent to performing statistical shrinkage.  They show that constraining portfolio weights to be nonnegative is equivalent to using the sample covariance matrix after reducing its large elements and then form the optimal portfolio without any restrictions on portfolio weights (that is, shrinking the covariance matrix).

# Building the Global Minimum Variance Portfolio

We've previously written the code to build the GMV portfolio, given the covariance. Let's start with just that:

```python
def sample_cov(r, **kwargs):
    """
    Returns the sample covariance of the supplied returns
    """
    return r.cov()

def weight_gmv(r, cov_estimator=sample_cov, **kwargs):
    """
    Produces the weights of the GMV portfolio given a covariance matrix of the returns 
    """
    est_cov = cov_estimator(r, **kwargs)
    return f.gmv(est_cov)
```

Now, let's run this new backtest. This may take a fair amount of time to run because we are running a backtest over nearly 50 years and both optimizing and rebalancing on a monthly basis.

```python
mv_s_r = f.backtest_ws(ind_rets, estimation_window=36, weighting=weight_gmv, cov_estimator=sample_cov)
btr = pd.DataFrame({"EW": ewr, "CW": cwr, "GMV-Sample": mv_s_r})
f.summary_stats(btr.dropna())
```

```python
(1+btr).dropna().cumprod().plot(figsize=(22,9), title="Industry Portfolios", grid = True)
```

Now, let's try a new estimator - Constant Correlation. The idea is simple, take the sample correlation matrix, compute the average correlation and then reconstruct the covariance matrix. The relation between correlations $\rho$ and covariance $\sigma$ is given by:

$$ \rho_{ij} = \frac{ \sigma_{ij} } {\sigma_{ii}  \sigma_{jj}} $$


```python
def cc_cov(r, **kwargs):
    """
    Estimates a covariance matrix by using the Elton/Gruber Constant Correlation model
    """
    rhos = r.corr()
    n = rhos.shape[0]
    # this is a symmetric matrix with diagonals all 1 - so the mean correlation is ...
    rho_bar = (rhos.values.sum()-n)/(n*(n-1))
    ccor = np.full_like(rhos, rho_bar)
    np.fill_diagonal(ccor, 1.)
    sd = r.std()
    ccov = ccor * np.outer(sd, sd)
    
    return pd.DataFrame(ccov, index=r.columns, columns=r.columns)
```

```python
wts = pd.DataFrame({
    "EW": f.weight_ew(ind_rets["2016":]),
    "CW": f.weight_cw(ind_rets["2016":], cap_weights=ind_mcap),
    "GMV-Sample": weight_gmv(ind_rets["2016":], cov_estimator=sample_cov),
    "GMV-ConstCorr": weight_gmv(ind_rets["2016":], cov_estimator=cc_cov),
})
```

```python
wts.T.plot.bar(stacked=True, figsize=(22,9), legend=False, grid = True);
```

```python
mv_cc_r = f.backtest_ws(ind_rets, estimation_window=36, weighting=weight_gmv, cov_estimator=cc_cov)
btr = pd.DataFrame({"EW": ewr, "CW": cwr, "GMV-Sample": mv_s_r, "GMV-CC": mv_cc_r})
f.summary_stats(btr.dropna())
```

```python
(1+btr).cumprod().dropna().plot(figsize=(22,9), title="Industry Portfolios", grid = True)
```

At least in this case, the CC Covariance Matrix did not enhanced the portfolios. One reason for this is that we are dealing with industries, so much of the noise of the individual stocks has already been wiped out.

### Statistical Shrinkage

We can mix the model and sample estimates by choosing a shrinkage parameter. You can either let the numbers dictate an optimal shrinkage value for $\delta$ although in practice many practiotioners choose $0.5$. Let's implement a simple shrinkage based covariance estimator that shrinks towards the Constant Correlation estimate.

```python
def shrinkage_cov(r, delta=0.5, **kwargs):
    """
    Covariance estimator that shrinks between the Sample Covariance and the Constant Correlation Estimators
    """
    prior = cc_cov(r, **kwargs)
    sample = sample_cov(r, **kwargs)
    return delta*prior + (1-delta)*sample
```

```python
wts = pd.DataFrame({
    "EW": f.weight_ew(ind_rets["2013":]),
    "CW": f.weight_cw(ind_rets["2013":], cap_weights=ind_mcap),
    "GMV-Sample": weight_gmv(ind_rets["2013":], cov_estimator=sample_cov),
    "GMV-ConstCorr": weight_gmv(ind_rets["2013":], cov_estimator=cc_cov),
    "GMV-Shrink 0.5": weight_gmv(ind_rets["2013":], cov_estimator=shrinkage_cov),
})
```

```python
wts.T.plot.bar(stacked=True, figsize=(22,9), legend=False, grid = True)
```

```python
mv_sh_r = f.backtest_ws(ind_rets, estimation_window=36, weighting=weight_gmv, cov_estimator=shrinkage_cov, delta=0.5)
btr = pd.DataFrame({"EW": ewr, "CW": cwr, "GMV-Sample": mv_s_r, "GMV-CC": mv_cc_r, 'GMV-Shrink 0.5': mv_sh_r})
(1+btr).cumprod().dropna().plot(figsize=(22,9), grid =True, title="49 Industry Portfolios")
f.summary_stats(btr.dropna())
```

# 2. The curse of non-stationarity

The underlying assumption when estimating returns parameters is that these parameters are constant. Nonetheless, it is well documented that volatility is not constant over time. Time-varying parameters can be handled using rolling windows, instead of the whole sample period. This helps to address the problem, but we still have a backward-looking bias because we are using the past sample data to estimate volatility today. 

That is why we need to introduce new methodologies that can be used to give us more meaningful, more forward-looking or current-looking estimates of volatility when we have a concern that volatility may be time-varying, which is most often the case. 


## 1. Exponentially Weighted Estimate

One simple way to improve the rolling window estimate, is to find a way to give more importance to more recent values, in order to make ir more current-looking. An efficient and simple method for doing this is to compute exponentially weighted estimates.

If we assume zero expected returns, then the variance of an asset return reduces to:

$$\hat{\sigma}_T^2 = \frac{1}{N}\sum R_t^2$$

Instead of assigning equal weights to all the observations, we can set:

$$ \hat{\sigma}_T^2 =\sum  \alpha_t R_t^2 $$

where  $\sum \alpha_t = 1$

And we design the value of $\alpha$ such as it decreases as the observation is further away in time.

In the Expoential Weighted Moving Average (EWMA) sets:

$$\alpha_t = \frac{\lambda^{T-t}}{\sum_{t=0}^T \lambda^{T-t}}$$

where $\lambda$ is the decay coefficient. The lower $\lambda$, the faster the importance of further away in time observations goes to zero.


## 2. Conditional Heteroscedastic Models

Conditional heteroscedastic models are econometric tools for modeling volatility over time. ARCH and GARCH models have proven to be extremely powerful methodologies for estimating time-varying risk parameters. 


## ARCH models

The first model that provides a systematic framework for volatility modeling is the ARCH model of Engle (1982). The basic idea of ARCH models is that (a) the shock $\epsilon_t$ of an asset return is serially uncorrelated, but dependent, and (b) the dependence of $\epsilon_t$ can be describer by a simple quadratic function of its lagged values. Specifically, an $ARCH(m)$ model assumes that:

$$\epsilon = \sigma_tz_t$$

$$\sigma_t^2 = \omega + \alpha_1 \epsilon_{t-1}^2 + \dots + \alpha_m\epsilon_{t-m}^2$$

Where $z_t$ is a sequence of white noise, $\omega > 0$ and $\alpha_i \geq 0$ fot $i>0$.

From the structure of the model, it is seen that large past squared shocks $\lbrace\epsilon_{t-i}^2 \rbrace_{i=1}^m$ imply a large conditional variance $\sigma_t^2$ for the innovation $\epsilon_t$.

### ARCH (1)

$$\epsilon = \sigma_tz_t$$

$$\sigma_t^2 = \omega + \alpha_1 \epsilon_{t-1}^2 $$

The unconditional mean of $\epsilon_t$ is zero becuase:

$$E(\epsilon_t) = E[E(\epsilon_t|F_{t-1})] = E[\sigma_tE(z_t)] = 0$$

The unconditional variance can be obtained as

$$Var(\epsilon_t) = E(\epsilon_t^2) = E[E(\epsilon_t^2|F_{t-1})] = E(\omega_0 + \alpha_1\epsilon_{t-1}^2) = \omega + \alpha_1 E(\epsilon_{t-1}^2)$$

Because $\epsilon_t$ is stationary with $E(\epsilon_t) = 0$, then $Var(\epsilon_t) = Var(\epsilon_{t-1}) = E(\epsilon_{t-1}^2)$

Thus, $Var(\epsilon_{t}) = \dfrac{\omega}{(1 - \alpha_1)}$



## GARCH models

Although the ARCH model is simple, it often requires many parameters to adequately describe the volatility process of an asset return. An extension of this model is known as the generalized ARCH (**GARCH**) model. 

A $GARH(m,s)$ model is defined as:

$$\epsilon_t = \sigma_tz_t$$

$$\sigma_t^2 = \omega + \sum_{i=1}^m\alpha_i\epsilon_{t-i}^2 + \sum_{j=1}^s\beta_j\sigma_{t-j}^2$$

Where $\omega >0$, $\alpha_i \geq 0$, $\beta_i \geq 0$ and $\sum (\alpha_i + \beta_i \leq 1$). $z_t$ is assumed to follow a standard normal or standardized Student-t distriburion. 


### GARCH(1,1)


$$\epsilon_t = \sigma_t z_t $$

$$\sigma_t^2 = \omega + \alpha_1 \epsilon_{t-1}^2+\beta_1 \sigma_{t-1}^2 $$

The uncionditional variance of $\epsilon$ is

$$Var(\epsilon) = \dfrac{\omega}{1 - \alpha_1 - \beta_1}$$


**Warning** Using ARCH and GARCH models is really useful, but it increases the problem of estimating many parameters.
